
from constraint import Problem, AllDifferentConstraint

# Define the classroom size (3x3 grid)
classroom_seats = {
    1: "Front-Left", 2: "Front-Center", 3: "Front-Right",
    4: "Middle-Left", 5: "Middle-Center", 6: "Middle-Right",
    7: "Back-Left", 8: "Back-Center", 9: "Back-Right"
}

# Define students
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Fiona', 'George', 'Helen', 'Ian']

# Create a CSP problem instance
problem = Problem()

# Add variables (students) with their possible domains (seats)
problem.addVariables(students, classroom_seats.keys())

# Add constraints
# All students must sit in different seats
problem.addConstraint(AllDifferentConstraint())

# Some students want to sit next to each other (Bob and Charlie)
def near_seats(a, b):
    adjacent_pairs = [
        (1, 2), (2, 3), (4, 5), (5, 6), (7, 8), (8, 9),  # Horizontal neighbors
        (1, 4), (4, 7), (2, 5), (5, 8), (3, 6), (6, 9)   # Vertical neighbors
    ]
    return (a, b) in adjacent_pairs or (b, a) in adjacent_pairs

problem.addConstraint(near_seats, ('Bob', 'Charlie'))

# Some students must sit far apart (Alice and David)
def far_seats(a, b):
    adjacent_pairs = [
        (1, 2), (2, 3), (4, 5), (5, 6), (7, 8), (8, 9),
        (1, 4), (4, 7), (2, 5), (5, 8), (3, 6), (6, 9)
    ]
    return (a, b) not in adjacent_pairs and (b, a) not in adjacent_pairs

problem.addConstraint(far_seats, ('Alice', 'David'))

# Some students prefer sitting in front of the class (Eva, Fiona)
def front_seats(seat):
    return seat in [1, 2, 3]  # Front row seats

problem.addConstraint(front_seats, ['Eva', 'Fiona'])

# Some students prefer sitting in the back (George, Helen)
def back_seats(seat):
    return seat in [7, 8, 9]  # Back row seats

problem.addConstraint(back_seats, ['George', 'Helen'])

# Solve the problem
solution = problem.getSolution()

# Print the solution
if solution:
    print("Seating arrangement:")
    for student, seat in solution.items():
        print(f"{student} will sit in {classroom_seats[seat]}")
else:
    print("No solution found.")


```python
# Function to print the board
def print_board(board):
    for row in board:
        print(" ".join(str(cell) for cell in row))

# Check if a number can be placed at the given position
def is_valid(board, row, col, num):
    # Check the row
    for i in range(9):
        if board[row][i] == num:
            return False

    # Check the column
    for i in range(9):
        if board[i][col] == num:
            return False

    # Check the 3x3 grid
    start_row = row - row % 3
    start_col = col - col % 3
    for i in range(3):
        for j in range(3):
            if board[start_row + i][start_col + j] == num:
                return False

    return True

# Function to solve the Sudoku puzzle using backtracking
def solve_sudoku(board):
    empty_cell = find_empty(board)
    if not empty_cell:
        # No empty space left, puzzle solved
        return True
    else:
        row, col = empty_cell

    for num in range(1, 10):
        if is_valid(board, row, col, num):
            board[row][col] = num

            # Recur to attempt to solve further
            if solve_sudoku(board):
                return True

            # Undo the move (backtrack)
            board[row][col] = 0

    return False

# Function to find an empty space in the Sudoku grid (denoted by 0)
def find_empty(board):
    for row in range(9):
        for col in range(9):
            if board[row][col] == 0:
                return (row, col)  # Return the row, col of the empty cell
    return None

# Example Sudoku puzzle (0 denotes an empty cell)
sudoku_board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

# Solve the Sudoku puzzle
if solve_sudoku(sudoku_board):
    print("Solved Sudoku:")
    print_board(sudoku_board)
else:
    print("No solution exists")

```
